1. Read the three CSV files which contain the score of the same students in term1
of each Subject
2. Remove the name and ethnicity column (to ensure confidentiality)
3. Fill missing score data with zero
4. Merge the three files
5. Change Sex(M/F) Column to 1/2 for further analysis
6. Store the data in a new file â€“ ScoreFinal.csv

```python
import pandas as pd
p_score = pd.read_csv('PhysicsScoreTerm1.csv')
m_score = pd.read_csv('MathScoreTerm1.csv')
d_score = pd.read_csv('DSScoreTerm1.csv')
```

```python
print(p_score.describe(), '\n', m_score.describe(), '\n', d_score.describe())
```

```python
m_score.isnull().sum()
```

```python
p_score['Score'] = p_score['Score'].fillna(0)
m_score['Score'] = p_score['Score'].fillna(0)
d_score['Score'] = p_score['Score'].fillna(0)
```

```python
m_score.head()
```

```python
df_mp = m_score.merge(p_score, how='left', on=['Name', 'Age', 'Ethinicity', 'Sex', 'ID'], suffixes=('_maths', '_physics'))
```

```python
df_mpd = df_mp.merge(d_score, how='left', on=['Name', 'Age', 'Ethinicity', 'Sex', 'ID'])
```

```python
df_mpd
```

```python
df_mpd.drop(['Name', 'Ethinicity', 'Subject_maths', 'Subject_physics', 'Subject'], axis=1, inplace=True)
```

```python
df_mpd.rename(columns={'Score': 'Score_datastructure'}, inplace=True)
```

```python
df_mpd.isnull().sum()
```

```python
df_mpd['Sex'].replace({'M': 1, 'F': 2}, inplace=True)
```

```python
df_mpd
```

```python
df_mpd.to_csv('ScoreFinal.csv')
```
