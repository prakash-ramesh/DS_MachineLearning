```python
import pandas as pd
df = pd.read_csv('BigMartSalesData.csv')
df.head()
```

```python
df_11 = df[df['Year'] == 2011]
```

```python
monthly_sales = df_11.groupby('Month')['Amount'].sum()
```

```python
monthly_sales.plot()
```

```python
monthly_sales.plot.bar()
```

```python
country_sales = df_11.groupby('Country')['Amount'].sum()
```

```python
import matplotlib.pyplot as plt

countries = country_sales.keys().to_list()
```

```python
plt.figure(figsize=(10,10))
plt.pie(country_sales.values, labels = countries, labeldistance=1)
plt.legend(loc='right')
plt.show()
```
